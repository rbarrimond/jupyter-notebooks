Initial exploratory analysis of the data published from [Carto](https://phl.carto.com/).
<hr />

Philadelphia uses CARTO so setup and query the data we need and store in a Panda dataframe.

```python
!pip3 install carto
import carto
import pandas as pd

from carto.auth import BaseAuthClient
auth_client = BaseAuthClient(base_url="https://phl.carto.com/")

from carto.sql import SQLClient
sql = SQLClient(auth_client)

try:
    data = sql.send("SELECT * FROM incidents_part1_part2 WHERE dispatch_date LIKE '2019-05-%'")
except CartoException as e:
    print("some error ocurred", e)

print(data['rows'])
```
